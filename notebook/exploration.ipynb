```python
import pandas as pd

samples = pd.read_csv("../data/s_MTBLS2550.txt", sep="\t")

print(samples.shape)
samples.head(1)
```

```python
samples.columns
```

```python
metabolites = pd.read_csv(
    "../data/m_MTBLS2550_NMR_CrAdj_metabolite_profiling_v2_maf.tsv",
    sep="\t"
)

print(metabolites.shape)
metabolites.head()
```

```python
metabolites.columns.value_counts
```

```python
cols_importantes = [
    'Source Name',
    'Factor Value[Sex]',
    'Factor Value[Tumor state]',
    'Factor Value[Tumor grade]',
    'Factor Value[Age]',
    'Factor Value[Neutral Status]',
    'Factor Value[ER status]',
    'Factor Value[HER2 status]',
    'Factor Value[Intercurrent tumors]',
    'Factor Value[Breed]'
]
```

```python
samples_clean = samples[cols_importantes]
```

```python
samples_clean.head(10)
```

```python
samples_clean = samples_clean.rename(columns={
    'Source Name': 'sample_id',
    'Factor Value[Sex]': 'sex',
    'Factor Value[Tumor state]': 'tumor_state',
    'Factor Value[Tumor grade]': 'tumor_grade',
    'Factor Value[Age]': 'age',
    'Factor Value[Neutral Status]': 'neutered',
    'Factor Value[ER status]': 'er_status',
    'Factor Value[HER2 status]': 'her2_status',
    'Factor Value[Intercurrent tumors]': 'tumor_type',
    'Factor Value[Breed]': 'breed'
})
```

```python
samples_clean.head(10)
```

```python
for col in metabolites.columns:
    print(col)
```

```python
metabolites_ml = metabolites[['metabolite_identification']].copy()
```

```python
sample_cols = [c for c in metabolites.columns if "_CrAdj" in c]

metabolites_ml = metabolites[['metabolite_identification'] + sample_cols]
```

```python
metabolites_ml = metabolites_ml.set_index('metabolite_identification')
```

```python
metabolites_ml = metabolites_ml.T
```

```python
metabolites_ml = metabolites_ml.reset_index()
metabolites_ml = metabolites_ml.rename(columns={'index': 'sample_id'})
```

```python
metabolites_ml['sample_id'] = metabolites_ml['sample_id'].str.replace('_CrAdj','')
```

```python
dataset_final = pd.merge(
    samples_clean,
    metabolites_ml,
    on="sample_id",
    how="inner"
)
```
